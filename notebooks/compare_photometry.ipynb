# Compare photometry from two MER runs

Takes two output MER catalogs and compare the photometry

### Imports

```python
from Catalogs import *
import os
```

### Configuration

```python
data_path = '/home/lmaurin/Work/PF_MER/data/Catalogs'
filename_1 = 'run_1/EUC_MER_FINAL-CAT_78C3FC_20200218T183832.206926Z_00.00.fits'
filename_2 = 'run_2/EUC_MER_FINAL-CAT_BA775E_20200225T183250.843496Z_00.00.fits'
```

### Load and merge catalogs

```python
cat1 = MERCatalog.from_file(os.path.join(data_path,filename_1))
cat2 = MERCatalog.from_file(os.path.join(data_path,filename_2))

mcat = MatchedCatalog.from_catalogs(cat1, cat2)
```

### Plots

```python
plot_tools.compare_mer_flux_outputs(mcat)



```
